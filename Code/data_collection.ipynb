```python
import time
from utility import Market, download_and_store, ALL_TICKERS

date_str = "2026-04-06" 

all_tickers = (
    [(tic, Market.US) for tic in ALL_TICKERS]
)
 
for i, (tic, market) in enumerate(all_tickers, start=1):
    download_and_store(tic, date_str, market)

    # yfinance accepts 60 calls every 60s, I currently have 
    # 4 intervals, so 15 calls per minute
    if i % 15 == 0:
        print(f"Throttling: completed {i} downloads, sleeping...")
        time.sleep(61)
```
